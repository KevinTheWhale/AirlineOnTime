```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

# create dictionary mapping airline codes to numbers
airline_dict = {'AA': 1, 'AS': 2, 'B6': 3, 'DL': 4, 'F9': 5, 'G4': 6, 'HA': 7, 'NK': 8, 'UA': 9, 'WN': 10}

# replace airline codes with corresponding numbers
dec_data['MKT_UNIQUE_CARRIER'] = dec_data['MKT_CARRIER'].replace(airline_dict)

# select predictors and target variable
X = dec_data[['MKT_UNIQUE_CARRIER', 'DISTANCE', 'DEP_DELAY_NEW']]
y = dec_data['ARR_DELAY_NEW']

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# create and fit decision tree model
tree_model = DecisionTreeRegressor(max_depth= 5)
tree_model.fit(X_train, y_train)

# predict ARR_DELAY_NEW for specific input values
input_data = [[1, 300, 10]]
predicted_delay = tree_model.predict(input_data)

print("Predicted ARR_DELAY_NEW:", predicted_delay)
```
